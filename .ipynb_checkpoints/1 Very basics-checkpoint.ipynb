```python
import numpy as np
import plotly.graph_objects as go
```

### Problem:
In machine learning and statistical problems, we have a set of known data $D$, unknown data $X$, and often model parameters $\theta$. <br>
$D$ and $X$ often come in the form of input and output data, but for simple cases, they are just a set of events.
Ideally, we would know $p(X|D)$. This is a non-parameteric model, as there is no $\theta$ involved. However, these models are often computationally intractable for large datasets. <br>
An example of a non-parameteric model is the K-nearest neighbours algorithm. $K$ is a hyper-parameter, and the probability of the data belonging to a class is given: $$p(y=c|x,D,K)=\frac{1}{K}\sum_{i} (i==c)$$

```python
X_D = np.random.rand(100,2)
Class_D = (X_D[:,0]<0.5)*1+(X_D[:,1]<0.75)*1 # fake data

trace1 = go.Scatter(x=X_D[Class_D==0,0],y=X_D[Class_D==0,1],mode="markers",marker=dict(color="red",size=6))
trace2 = go.Scatter(x=X_D[Class_D==1,0],y=X_D[Class_D==1,1],mode="markers",marker=dict(color="blue",size=6))
trace3 = go.Scatter(x=X_D[Class_D==2,0],y=X_D[Class_D==2,1],mode="markers",marker=dict(color="green",size=6))

layout = go.Layout(
    title = "KNN",
    xaxis = dict(
        title = 'X1',
    ),
    yaxis = dict(
        title = 'X2',
    ),
    font=dict(size=16),
    showlegend=False,
    margin=dict(t=40,l=10,b=10,r=10),
    legend=dict(
        orientation="v",
        yanchor="top",
        y=0.98,
        xanchor="left",
        x=0.02
    )
)
fig = go.Figure(data=[trace1,trace2,trace3],layout=layout)
fig.show(renderer="png",width=300,height=300)
```
